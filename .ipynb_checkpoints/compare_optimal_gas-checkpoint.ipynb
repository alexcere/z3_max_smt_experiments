```python
results_files = ["model_found.csv", "no_model_found.csv", "no_model_found_oms.csv", "no_model_found_z3.csv"]

def extract_values(l):
    # return [elem for elem in l.strip('][').split(', ')]
    return list(filter(lambda x: x != '', l.strip('][').split(', ')))

def select_nearest_index_val(timeout, found_times):
    if found_times == []:
        return pd.NA
    idx = (np.abs(np.array(found_times) - timeout)).argmin()
    if idx == len(found_times) - 1 or found_times[idx+1] - timeout >= 6:
        return idx
    else:
        return idx + 1

def total_instr(instructions_disasm):
    if pd.isna(instructions_disasm):
        return pd.NA
    instructions = list(filter(lambda x: x != '', instructions_disasm.split(' ')))
    i, j = 0, 0
    while i < len(instructions):
        j += 1
        if instructions[i] == "PUSH":
            i += 2
        elif instructions[i].startswith("PUSH") and not instructions[i].startswith("PUSHDEPLOYADDRESS") \
                    and not instructions[i].startswith("PUSHSIZE"):
            i += 2
        else:
            i += 1
    return j
```

```python
cv_file = "model_found.csv"
smt_folder = "smt_encoding_model_found/"
df = pd.read_csv(cv_file).to_dict('r')
smt_files = list(glob.glob(smt_folder + "*z3.smt2"))

# Problem: rstrip seems to erase block number in some cases. The order among
# elements in the df and the glob match, so we build a dict to link them
names = dict()

assert len(df) == len(smt_files)
for row, smt_name in zip(df, smt_files):
    file_name = smt_name.split("/")[-1]
    assert file_name.startswith(row["block_id"]), file_name + " " + row["block_id"]
    names[row["block_id"]] = file_name
```

```python
df = pd.read_csv(cv_file).to_dict('r')
new_rows = []
for elem in df:
    block_id = names[elem["block_id"]]
    contract_name = block_id.split("_")[0]
    
    z3_file = list(glob.glob(folder + "/z3_10s/"  + contract_name  + "*.csv"))[0]
    oms_file = list(glob.glob(folder + "/oms_10s/"  + contract_name  + "*.csv"))[0]
    
    z3_rows = pd.read_csv(z3_file).to_dict('r')
    oms_rows = pd.read_csv(oms_file).to_dict('r')
    
    selected_z3_row = list(filter(lambda x: x["block_id"] + "_z3.smt2" == block_id, z3_rows))[0]
    selected_oms_row = list(filter(lambda x: x["block_id"] + "_z3.smt2" == block_id, oms_rows))[0]
    
    new_row = {"block_id": block_id}
    
    models_cost = list(map(lambda x: int(x), extract_values(elem["cost_per_model_found"])))
    models_time = list(map(lambda x: float(x), extract_values(elem["found_model_times"])))
    
    timeout = elem["timeout_in_s"]
    
    if elem["finished_execution"]:
        best_model_time = elem["solver_time_in_sec"]
        best_model_cost = models_cost[-1]
    
    else:
        idx = select_nearest_index_val(timeout, models_time)
        if pd.isna(idx):
            best_model_time = pd.NA
            best_model_cost = pd.NA
        
        else:
            best_model_time = max(models_time[idx], timeout)
            best_model_cost = models_cost[idx]
            
    new_row["timeout"] = timeout
    
    new_row["hitting_set_time"] = best_model_time
    new_row["hitting_set_cost"] = best_model_cost
    new_row["hitting_set_found_optimal"] = elem["finished_execution"]
    
    new_row["hard_time"] = elem["hard_time_in_sec"]
    new_row["hard_cost"] = elem["hard_cost_found"]
    
    new_row["oms_time"] = selected_oms_row["solver_time_in_sec"]
    new_row["oms_cost"] = selected_oms_row["init_progr_len"] - total_instr(selected_oms_row["target_disasm"])
    new_row["oms_found_optimal"] = selected_oms_row["shown_optimal"]
    
    new_row["z3_time"] = selected_z3_row["solver_time_in_sec"]
    new_row["z3_cost"] = selected_z3_row["init_progr_len"] - total_instr(selected_z3_row["target_disasm"])
    new_row["z3_found_optimal"] = selected_z3_row["shown_optimal"]
    
    # if new_row["oms_found_optimal"] and pd.notnull(new_row["hitting_set_cost"]) and new_row["hitting_set_cost"] > new_row["oms_cost"]:
    new_rows.append(new_row)
    
df_new = pd.DataFrame(new_rows)
df_new.to_csv(cv_file.rstrip(".csv") + "_times_cost.csv")
```

```python
df = pd.read_csv("statistics.csv").to_dict('r')
new_rows = []
for elem in df:
    block_id = elem["block_id"]
    contract_name = block_id.split("_")[0]
    
    z3_file = list(glob.glob(folder + "/z3_10s/"  + contract_name  + "*.csv"))[0]
    oms_file = list(glob.glob(folder + "/oms_10s/"  + contract_name  + "*.csv"))[0]
    
    z3_rows = pd.read_csv(z3_file).to_dict('r')
    oms_rows = pd.read_csv(oms_file).to_dict('r')
    
    selected_z3_row = list(filter(lambda x: x["block_id"].startswith(block_id), z3_rows))[0]
    selected_oms_row = list(filter(lambda x: x["block_id"].startswith(block_id), oms_rows))[0]
    
    new_row = deepcopy(elem)
    
    new_row["solver_time_in_sec_z3"] = selected_z3_row["solver_time_in_sec"]
    new_row["solver_time_in_sec_z3%"] = round(elem["solver_time_in_sec"] / selected_z3_row["solver_time_in_sec"],3)
    
    new_row["solver_time_in_sec_oms"] = selected_oms_row["solver_time_in_sec"]
    new_row["solver_time_in_sec_oms%"] = round(elem["solver_time_in_sec"] / selected_oms_row["solver_time_in_sec"],3)
    
    if new_row["solver_time_in_sec_z3"] >= 5:
        new_rows.append(new_row)
    
df_new = pd.DataFrame(new_rows)
df_new.to_csv("combined_statistics.csv")
```
